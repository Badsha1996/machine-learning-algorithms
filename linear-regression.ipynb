#                                                                    LINEAR REGRESSION


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression as LR

```python
df = pd.read_csv("data/age-weight-RL-data.csv")
df =  df.iloc[:-2]
df
```

```python
%matplotlib inline
plt.xlabel("Weight(Kg)")
plt.ylabel("Age(years)")
plt.title("Linear Regression of AGE vs Weight")
plt.scatter(df["age"], df["weight"])
plt.show()
```

```python
X = df[["weight"]] # Feature # 2D
y = df["age"]      # the value i want to predict  # 1D
```

```python
# Creating a model/ machine learning Model
my_RL_model = LR() # Class -> Object
my_RL_model.fit(X, y)  # Taining the model
```

```python
y_pred = my_RL_model.predict([[60]])
```

```python
df = pd.read_csv("data/age-weight-RL-predict.csv")
df
```

```python
weight_predict = my_RL_model.predict(df[["weight"]])
df["age"] = weight_predict
df["age"] = df["age"].astype(int)
df.to_csv("data/output/model-prediction.csv", index=False)
df
```

```python
%matplotlib inline 
plt.xlabel("Weight(Kg)")
plt.ylabel("Age(years)")
plt.title("Linear Regression of AGE vs Weight")
plt.scatter( df["weight"], df["age"])
plt.plot(df[["weight"]],weight_predict, color="red")
```
