```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv("./data/sat-admit-data.csv")
df
```

```python
df.isna().any() # To check whether there is a missing value 
```

```python
df["Admitted"] = df["Admitted"].map({"Yes" : 1, "No" : 0})
df
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
plt.scatter(df["SAT"], df["Admitted"])
plt.xlabel("SAT")
plt.ylabel("Admitted")
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression as LR
from sklearn.model_selection import train_test_split as tts
```

```python
X = df[["SAT"]]
y = df["Admitted"]
X_train, X_test, y_train, y_test = tts(X, y, test_size=0.2)
```

```python
model = LR()
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
y_predict
```

```python
model.score(X_test, y_test)
```

```python
%matplotlib inline 
plt.scatter(df["SAT"], df["Admitted"])
plt.xlabel("SAT")
plt.ylabel("Admitted")

x_plt = np.sort(df[["SAT"]].values, axis= 0)
y_plt = model.predict_proba(x_plt)[:,1]
plt.plot(x_plt, y_plt)

plt.show()
```

```python
import joblib
joblib.dump(model, "./saved_models/logesticModel.joblib")
```
